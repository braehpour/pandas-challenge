```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
```

```python
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data_df['SN'].unique())
total_players_df = pd.DataFrame({'Total Players': [total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
item_count = purchase_data_df['Item ID'].nunique()
#https://stackoverflow.com/questions/38309729/count-unique-values-with-pandas-per-groups

avg_price = round(purchase_data_df['Price'].mean(),2)
total_purchases = len(purchase_data_df['Purchase ID'])
total_revenue = purchase_data_df['Price'].sum()
```

```python
calculations_df =pd.DataFrame({'Number of Unique Items': [item_count],
       'Average Price': [avg_price],
       'Number of Purchases': [total_purchases],
       'Total Revenue': [total_revenue]})
calculations_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_grouped = purchase_data_df.groupby(['Gender'])['SN'].nunique()
gender_df = pd.DataFrame([gender_grouped]).transpose()    #https://note.nkmk.me/en/python-pandas-t-transpose/
gender_df.columns = ['Total Count']
gender_percent = round((gender_grouped/total_players)*100,2)
gender_df["Percentage of Players"] = gender_percent
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_grouped2 = purchase_data_df.groupby(['Gender'])['SN'].count()
gender_purchase_df = pd.DataFrame([gender_grouped2]).transpose()
gender_purchase_df.columns = ['Purchase Count']

avg_purchase_price = purchase_data_df.groupby(['Gender'])['Price'].mean()
gender_purchase_df["Average Purchase Price"] = round(avg_purchase_price,2)

total_purchase_price = purchase_data_df.groupby(['Gender'])['Price'].sum()
gender_purchase_df["Total Purchase Value"] = round(total_purchase_price,2)

total_purchase_price = purchase_data_df.groupby(['Gender'])['Price'].sum()
gender_purchase_df["Average Total per Person"] = round(avg_purchase_price,2)

gender_purchase_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
bins = pd.cut(purchase_data_df['Age'], [0, 9, 14, 19, 24, 29, 34, 39, 1000], labels = group_names)


age_df = purchase_data_df.groupby(bins)['Age']
age_count = age_df.count()
age_percent = round((age_count/total_players)*100,2)

total_count_df = pd.DataFrame([age_count]).transpose()
total_count_df.columns = ["Total Count"]
total_count_df["Percentage of Players"] = age_percent
total_count_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
bins = pd.cut(purchase_data_df['Age'], [0, 9, 14, 19, 24, 29, 34, 39, 1000], labels = group_names)

age_df = purchase_data_df.groupby(bins)['Age']
age_count = age_df.count()

total_count_df = pd.DataFrame([age_count]).transpose()

total_count_df.columns = ["Total Count"]

avg_price_age = round(purchase_data_df.groupby([bins])['Price'].mean(),2)
total_count_df["Average Purchase Price"] = avg_price_age

total_value = purchase_data_df.groupby(bins)['Price'].sum()
total_count_df["Total Purchase Value"]  = total_value

total_purchase_price = purchase_data_df.groupby([bins])['Price'].sum()
total_count_df["Average Total per Person"] = round(avg_price_age,2)


total_count_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
names = purchase_data_df['SN'].unique()
purchase_count = purchase_data_df.groupby('Purchase ID')['SN'].nunique()

top_spenders_df = pd.DataFrame()
top_spenders_df['SN'] = names
top_spenders_df['Purchase Count'] = purchase_count

avg_price_top = round(purchase_data_df.groupby(['Purchase ID'])['Price'].mean(),2)
top_spenders_df["Average Purchase Price"] = avg_price_top

total_price_top = round(purchase_data_df.groupby(['Purchase ID'])['Price'].sum(),2)
top_spenders_df["Total Purchase Price"] = total_price_top

top_spenders_df.sort_values('Purchase Count', ascending = False)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items_df = pd.DataFrame()
purchase_count = purchase_data_df['Item ID'].value_counts()
item_id = purchase_data_df['Item ID']
item_name = purchase_data_df['Item Name']
purchase_count = purchase_count
item_price = purchase_data_df['Price']
total_value = item_price*purchase_count

popular_items_df["Item Id"] = item_id
popular_items_df["Item Name"] = item_name
popular_items_df["Purchase Count"] = purchase_count
popular_items_df["Item Price"] = item_price
popular_items_df["Total Purchase Value"] = total_value

popular_items_df.sort_values("Purchase Count", ascending = False).head()
```
